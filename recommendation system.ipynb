```python
data1=pd.read_csv('E:/PYTHON/datascience/data/tmdb_5000_movies.csv')
data2=pd.read_csv('E:/PYTHON/datascience/data/tmdb_5000_credits.csv')
```

```python
movie_data=data1.merge(data2, on='title')
movie_data.head()
```

```python
movie_data.info()
```

```python
movie_data.columns
```

```python
movie_data['new_col'] = range(1, len(movie_data) + 1)
movie_data=movie_data.reset_index()
movie_data.head(1)
```

```python
movie_data.head(1)
```

```python
import ast
from ast import literal_eval
cols=['genres','crew','cast','keywords']
for col in cols:
    movie_data[col]=movie_data[col].apply(literal_eval)
```

```python
def convert(x):
    L=[]
    for i in x:
        L.append(i['name'])
    return L
```

```python
movie_data['genres']=movie_data['genres'].apply(convert)
movie_data['keywords']=movie_data['keywords'].apply(convert)
```

```python
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan
```

```python
movie_data['director']=movie_data['crew'].apply(get_director)
```

```python
def convert_cast(x):
    if isinstance(x, list):
        names = [i['name'] for i in x]
        if len(names) > 3:
            names = names[:3]
        return names
    return []
```

```python
movie_data['cast']=movie_data['cast'].apply(convert_cast)
```

```python
movie_data.head(1)
```

```python
movie_data.info()
```

```python
def convert_str(x):
    if isinstance(x, list):
        return [str.lower(i.replace(" ", "")) for i in x]
    else:
        if isinstance(x, str):
            return str.lower(x.replace(" ", ""))
        else:
            return ''
```

```python
spl_cols=['genres','cast','director','keywords']
for spl in spl_cols:
    movie_data[spl]=movie_data[spl].apply(convert_str)
```

```python
movie_data.head(1)
```

```python
def create_combine(x):
    return ' '.join(x['keywords']) + ' ' + ' '.join(x['cast']) + ' ' + x['director'] + ' ' + ' '.join(x['genres'])
movie_data['combine'] = movie_data.apply(create_combine, axis=1)
```

```python
movie_data['combine'].head(1)
```

```python
movie_data.info()
```

```python
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
vector=TfidfVectorizer(stop_words='english')
feature_vector=vector.fit_transform(movie_data['combine'])
```

```python
similariy = cosine_similarity(feature_vector,feature_vector)
```

```python
movie_name=input('enter movie name:')
```

```python
movie_name
```

```python
list_of_all_title=movie_data['title'].tolist()
```

```python
find_close_match=difflib.get_close_matches(movie_name, list_of_all_title)
find_close_match
```

```python
close_match=find_close_match[0]
close_match
```

```python
index_of_the_movie=movie_data[movie_data.title == close_match]['index'].values[0]
index_of_the_movie
```

```python
similariy_score=list(enumerate(similariy[index_of_the_movie]))
similariy_score
```

```python
sorted_similarity_movie=sorted(similariy_score, key= lambda x:x[1], reverse=True )
sorted_similarity_movie
```

```python
print('movie suggested for you: \n')

i = 1
for movie in sorted_similarity_movie:
    index = movie[0]
    title_from_index = movie_data[movie_data.index==index]['title'].values[0]
    if(i<=10):
        print(i,'.', title_from_index)
        i+=1
        
    
```
